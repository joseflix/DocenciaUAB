## <font color='green'>**> Exercici #1: El Joc del Caos i Fractals de Montecarlo [3.5 punts]**</font>

El "Joc del Caos" és un mètode matemàtic fascinant que utilitza la probabilitat i l'atzar per generar figures geomètriques altament ordenades, com ara fractals. En aquest exercici generarem el famós **Triangle de Sierpinski**.

Les regles del joc són extremadament senzilles:
1. Tenim els 3 vèrtexs d'un triangle equilàter.
2. Comencem en un punt $P$ qualsevol (pot ser l'origen o un punt aleatori).
3. A cada iteració, s'escull a l'atzar (amb la mateixa probabilitat) un dels 3 vèrtexs del triangle.
4. Ens movem exactament fins al **punt mitjà** entre la nostra posició actual $P$ i el vèrtex escollit. Aquesta serà la nostra nova posició $P$.
5. Anotem aquest nou punt i repetim el procés milers de vegades.

Sorprenentment, la col·lecció de tots els punts per on passem no serà una taca aleatòria, sinó un fractal perfecte.

1.1 **[0.25 punts]** Defineix en un array bidimensional els 3 vèrtexs d'un triangle equilàter de costat 1 recolzat sobre l'eix X. Les coordenades són: $V_0=(0, 0)$, $V_1=(1, 0)$ i $V_2=(0.5, \frac{\sqrt{3}}{2})$.

```python
# El teu codi

# 1. Definim els 3 vèrtexs del triangle equilàter
v_y_top = np.sqrt(3) / 2.0
vertexs = np.array([
    [0.0, 0.0],       # V0: Vèrtex inferior esquerre
    [1.0, 0.0],       # V1: Vèrtex inferior dret
    [0.5, v_y_top]    # V2: Vèrtex superior
])
```

1.2. **[0.25 punts]** Inicialitza un punt actual $P$ en coordenades $(0.5, 0.5)$ i prepara un array (o llistes) per guardar les coordenades $x$ i $y$ de **50.000 iteracions**.

```python
# El teu codi

# 2. Inicialització
N_passos = 50000
P_actual = np.array([0.5, 0.5]) # Punt de partida arbitrari

# Creem una matriu buida de Nx2 per guardar tots els punts per eficiència
# També es pot fer amb dues llistes x_vals = [] i y_vals = [] i fent .append()
punts_generats = np.zeros((N_passos, 2))
```

1.3 **[1.5 punts]** Escriu el bucle de generació. A cada pas, utilitza `np.random.randint` per triar quin vèrtex s'utilitza (0, 1 o 2), calcula el punt mitjà entre $P$ i el vèrtex, actualitza $P$ i guarda'l.

```python
# El teu codi

# 3. Bucle del Joc del Caos

for i in range(N_passos):
    # Triem un vèrtex a l'atzar (0, 1 o 2)
    idx_vertex = np.random.randint(0, 3)
    vertex_escollit = vertexs[idx_vertex]

    # Ens movem al punt mitjà entre la posició actual i el vèrtex escollit
    P_actual = (P_actual + vertex_escollit) / 2.0

    # Guardem la nova posició
    punts_generats[i] = P_actual
```

1.4 **[1.5 punts]** Fes un gràfic de dispersió (*scatter plot*) de tots els punts generats. Perquè el fractal es vegi bé, cal que el gruix dels punts sigui molt petit (utilitza el paràmetre `s=0.1` o `s=0.01` i `edgecolors='none'`). Posa la mateixa escala als dos eixos (`plt.axis('equal')`) perquè el triangle no es deformi.

```python
# El teu codi

# 4. Representació gràfica del fractal
plt.figure(figsize=(8, 8))

# Extraiem columnes d'X i Y per al scatter plot
# s=0.1 fa que els punts siguin petits i el fractal quedi ben dibuixat
plt.scatter(punts_generats[:, 0], punts_generats[:, 1], s=0.1, color='darkblue', alpha=0.5)

# Opcional: Dibuixem els 3 vèrtexs originals en vermell per referència
plt.plot(vertexs[:, 0], vertexs[:, 1], 'ro', markersize=5)

plt.title("Triangle de Sierpinski generat pel Joc del Caos")
plt.axis('equal') # Molt important perquè sigui un triangle equilàter a la pantalla
plt.axis('off')   # Amaguem els eixos perquè quedi un gràfic més net
plt.tight_layout()
plt.show()
```

2.2 **[0.25 punts]** Defineix una funció de Python que calculi el valor de $\chi^2$ donats els paràmetres $a$ i $b$, i les dades experimentals. Només la funció.


```python
# El teu codi

# 2. Funció del model i del Chi-quadrat
def model(x, a, b):
    return a * x**2 + b

def chi_square(a, b, x, y, err):
    y_pred = model(x, a, b)
    return np.sum(((y - y_pred) / err)**2)
```

2.3 **[1.5 punts]** Realitza un *grid search* (escaneig) bidimensional per trobar els valors d'$a$ i $b$ que minimitzen el $\chi^2$. Suposa que $a \in [2.0, 6.0]$ i $b \in [-5.0, 5.0]$, i utilitza 200 passos equidistants per a cada paràmetre. Imprimeix per pantalla l'$a_{òptim}$, el $b_{òptim}$ i el $\chi^2_{mínim}$.


```python
# El teu codi

# 3. Grid Search per trobar el mínim
A_scan = np.linspace(2.0, 6.0, 200)
B_scan = np.linspace(-5.0, 5.0, 200)

min_chi2 = np.inf
opt_a = None
opt_b = None

# Recorrem tot l'espai de paràmetres (doble bucle)
for a in A_scan:
    for b in B_scan:
        c2 = chi_square(a, b, x_data, y_data, sigma_y)
        if c2 < min_chi2:
            min_chi2 = c2
            opt_a = a
            opt_b = b

print(f"Paràmetres òptims trobats: a = {opt_a:.3f}, b = {opt_b:.3f}")
print(f"Chi-quadrat mínim: {min_chi2:.2f}")
```

2.5 **[0.5 punts]** Segons la teoria estadística, els **residus normalitzats**, definits com $r_i = \frac{y_i - y_{model}(x_i)}{\sigma_i}$, haurien de seguir una distribució Normal Estàndard $\mathcal{N}(\mu=0, \sigma=1)$ si l'ajust és bo. Fes un histograma normalitzat d'aquests residus $r_i$ i superposa-hi la funció de densitat de probabilitat (PDF) d'una Gaussiana estàndard per verificar-ho visualment.

```python
# 5. Anàlisi dels residus
residuals = (y_data - model(x_data, opt_a, opt_b)) / sigma_y

plt.figure(figsize=(8, 5))
# Histograma (density=True és clau per comparar amb la PDF)
plt.hist(residuals, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Residus normalitzats')

# Distribució Gaussiana Estàndard de referència
x_pdf = np.linspace(-5, 5, 100)
y_pdf = ss.norm.pdf(x_pdf, loc=0, scale=1)
plt.plot(x_pdf, y_pdf, 'g--', lw=2, label=r'$\mathcal{N}(\mu=0, \sigma=1)$ teòrica')

plt.xlabel(r'Residu normalitzat $r_i$')
plt.ylabel(r'Densitat de probabilitat')
plt.title("Anàlisi estadística dels residus de l'ajust")
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
