La funci贸n de test que mejor se ajusta es aquella que est谩 'm谩s' cerca de tu funci贸n cumulativa de muestra de acuerdo con esta distancia $d_{max}$. Primero vamos a hacer una gr谩fica de la funci贸n cumulativa de la muestra: en este caso, para cada valor $_i$ de tu muestra, la funci贸n cumulativa incrementa en un factor 1/50, as铆 hasta llegar a un valor m谩ximo de 1. 

Tambi茅n vamos a dibujar 2 funciones cumulativas de test, para ver como se acercan a la funci贸n cumulativa de tu muestra. Considera dos gaussianas con valor medio $\mu$=-4.45 (valor obtenido antes) y $\sigma_1$=10 y $\sigma_2$=28.

```
b = np.sort(a)
cummuestra=np.linspace(1/len(b),1,len(b))

s1 = 10
s2 = 28
x = np.linspace(np.min(b),np.max(b),1000)
y1 = ss.norm.cdf(x,loc=np.mean(a),scale=s1)
y2 = ss.norm.cdf(x,loc=np.mean(a),scale=s2)
                 
plt.figure(figsize=(8, 8))
plt.plot(b,cummuestra,'-', lw=2, color='k', label='mostra')
plt.plot(x,y1,'r--', lw=1, label=r'$\sigma_1$') 
plt.plot(x,y2,'g--', lw=1, label=r'$\sigma_2$') 
plt.legend(loc='best')
plt.title('CDF', fontsize=25)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$F(x)$', fontsize=18)
plt.show()
```

### <font color='orange'>**> Python #2**</font>

Vamos a calcular ahora la $\sigma_{贸ptima}$ usando la funci贸n cumulativa de tu muestra y comparando con diferentes funciones cumulativas de test. Fija el valor medio de la distribuci贸n Gaussiana en $\mu_m$ y considera 1000 pasos equidistantes entre $\sigma_{inf}$=15 y $\sigma_{sup}$=45. En este caso, para cada valor $x_i$ de tu muestra, la funci贸n cumulativa te incrementa en un factor 1/50, y queda fijada. 

Haz una gr谩fica de los valores $d_{max}$ en el rango [$\sigma_{inf}$,$\sigma_{sup}$] y eval煤a cual es el $\sigma_{贸ptimo}$ (el que minimiza la funci贸n $d_{max}$). El valor $d_{max,i}$ se calcula para cada funci贸n de test $i$ como:

<br>

$$
d_{max,i} = max_{n=1}^{50}( |CDF_{muestra}(x_n) - CDF_{Gaussiana}(x_n, \mu_m, \sigma_i) | )
$$

<br>

```
minsigma = 15
maxsigma = 45

b = np.sort(a)
cummuestra=np.linspace(1/len(b),1,len(b))

dmax = []
sigstep = np.linspace(minsigma,maxsigma,1000)

for i in range(len(sigstep)):
  dmax.append(np.max(np.abs(cummuestra-ss.norm.cdf(b,loc=np.mean(a),scale=sigstep[i])))) # Varios pasos?

plt.figure(figsize=(7, 7))
plt.plot(sigstep,dmax)
plt.title('M茅todo Kolmogorov-Smirnov', fontsize=25)
plt.xlabel('$\sigma$', fontsize=18)
plt.ylabel('$d_{max}$', fontsize=18)
plt.axvline(sigstep[dmax.index(np.min(dmax))], linestyle='--', color='blue')
#plt.yscale('log')
plt.show()

print("\n Sigma_贸ptimo KS = {:3.2f} \n".format(sigstep[dmax.index(np.min(dmax))]))
```
